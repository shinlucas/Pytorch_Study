```python
m1 = torch.FloatTensor([[3,3]])
m2 = torch.FloatTensor([[2,2]])
print(m1+m2)
```

```python
m1 = torch.FloatTensor([[1,2]])
m2 = torch.FloatTensor([3])
print(m1+m2)
```

```python
m1 = torch.FloatTensor([[1,2]])
m2 = torch.FloatTensor([[3], [4]])
print(m1+m2)
```

```python
print()
print('-------------')
print('Mul vs Matmul')
print('-------------')
m1 = torch.FloatTensor([[1, 2], [3, 4]])
m2 = torch.FloatTensor([[1], [2]])
print('Shape of Matrix 1: ', m1.shape) # 2 x 2
print('Shape of Matrix 2: ', m2.shape) # 2 x 1
print(m1.matmul(m2)) # 2 x 1

m1 = torch.FloatTensor([[1, 2], [3, 4]])
m2 = torch.FloatTensor([[1], [2]])
print('Shape of Matrix 1: ', m1.shape) # 2 x 2
print('Shape of Matrix 2: ', m2.shape) # 2 x 1
print(m1 * m2) # 2 x 2
print(m1.mul(m2))
```

```python
# Type Casting
lt = torch.LongTensor([1,2,3,4])
print(lt)
print(lt.float())
```

```python
bt = torch.ByteTensor([True,False,False,True])
print(bt)
print(bt.long()) # 정수
print(bt.float()) # float
```

```python
bt2 = (lt==2)
print(bt2)
print(bt2.long()) # 정수
print(bt2.float()) # float
```

```python
# Concatenate
x = torch.FloatTensor([[1,2,9], [3,4,10]])
y = torch.FloatTensor([[5,6,11], [7,8,12]])

# dim이 늘어남
print(torch.cat([x,y],dim=0))
print(torch.cat([x,y],dim=1))
```

```python
 # Stacking

x = torch.FloatTensor([1, 4,7])
y = torch.FloatTensor([2, 5,8])
z = torch.FloatTensor([3, 9,6])

print(torch.stack([x, y, z]))
print(torch.stack([x, y, z], dim=1))
```

```python
print(torch.cat([x.unsqueeze(0), y.unsqueeze(0), z.unsqueeze(0)], dim=0))
```

```python
# Ones zeros
x = torch.FloatTensor([[0,1,2],[2,1,0]])
print(x)
```

```python
print(torch.ones_like(x))
print(torch.zeros_like(x))
```

```python
# In-place Operation
x = torch.FloatTensor([[1, 2], [3, 4]])

print(x.mul(2.))
print(x)
print(x.mul_(2.)) # _(언더바)를 넣으면 반영이 됨
print(x)
```
